```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import plotly.graph_objects as go
import plotly.io as pio
pio.templates.default = "plotly_dark"
import pandas as pd
import numpy as np
import requests
import json
import math
import seaborn as sns
sns.set()
```

```python
api = requests.Session()
response = api.get('https://data.covid19.go.id/public/api/update.json', timeout=5).json()
response['update']['harian']
```

```python
res = {'date':[], 'confirmed':[], 'recovered':[], 'death':[], 'active':[]}
for name_dict in response['update']['harian']:
    res['date'].append(name_dict.get('key_as_string'))
    res['confirmed'].append(name_dict.get('jumlah_positif_kum'))
    res['recovered'].append(name_dict.get('jumlah_sembuh_kum'))
    res['death'].append(name_dict.get('jumlah_meninggal_kum'))
    res['active'].append(name_dict.get('jumlah_dirawat_kum'))
df = pd.DataFrame(res)
df.to_csv("covid19.csv", index=False)
```

```python
text = open('covid19.csv', 'r')
text = ''.join([i for i in text])
text = text.replace("T00:00:00.000Z", "")
text = text.replace("value", "")
text = text.replace(":", "")
text = text.replace("'", "")
text = text.replace("}", "")
text = text.replace("{ ", "")

x = open('covid19.csv', 'w')
x.writelines(text)
x.close()
```

```python
full_data = pd.read_csv('covid19.csv', parse_dates=['date'])
full_data.head()
```

```python
data = full_data
data.describe()
```

```python
figure(num=None, figsize=(14, 10), dpi=80, facecolor='w', edgecolor='k')
plt.title('Confirmed & Active Cases in Indonesia')
plt.xlabel('Date')
plt.ylabel('# of Cases')

plt.plot(data.date, data.confirmed, '-', label='Confirmed')
plt.plot(data.date, data.active, '-', label='Active Cases')

plt.grid(alpha=0.2)
plt.legend(loc="upper left")
```

```python
dpsbb=data.date[53:]
cpsbb=data.confirmed[53:]

fig, ax = plt.subplots(figsize=(14,10))
plt.title('Number of Cases Since PSBB 24 Apr 2020')
sns.lineplot(x=dpsbb, y=cpsbb)
xticks=ax.xaxis.get_major_ticks()
for i in range(len(xticks)):
    if i%2==1:
        xticks[i].set_visible(False)
```

```python
n = 1
ncase=[0]
while n < len(data.confirmed):
    ncase.append(data.confirmed[n] - data.confirmed[n-1])
    n+=1
day = np.array(range(1,len(ncase)+1))
```

```python
figure(num=None, figsize=(14, 10), dpi=80, facecolor='w', edgecolor='k')

plt.title('New Cases in Indonesia Starting From 02-03-2020')
plt.xlabel('Day')
plt.ylabel('# of New Cases')

plt.axvline(53, color='gray', linestyle='--')
plt.text(x=53, y=2000, s=' PSBB', alpha=0.5, color='#334f8d')

plt.bar(day, ncase, label='New Cases', color='#4b8bbe')

plt.grid(alpha=0.8)
plt.legend(loc="upper left")
```

```python
figure(num=None, figsize=(14, 10), dpi=80, facecolor='w', edgecolor='k')

plt.title('Closed Cases in Indonesia')
plt.xlabel('Day')
plt.ylabel('# of Cases')

plt.plot(data.date, data.recovered, '-', label='Recovered', color='blue')
plt.plot(data.date, data.death, '-', label='Deaths', color='green')

plt.grid(alpha=0.2)
plt.legend(loc="upper left")
```

```python
daily_cases = []
for i in range(0, data.shape[0]):
    if i == 0:
        daily_cases.append(data.iloc[0]['confirmed'])
    else:
        daily_cases.append(data.iloc[i]['confirmed']-data.iloc[i-1]['confirmed'])
data['daily_cases']=daily_cases
```

```python
daily_recovered = []
for i in range(0, data.shape[0]):
    if i == 0:
        daily_recovered.append(data.iloc[0]['recovered'])
    else:
        daily_recovered.append(data.iloc[i]['recovered']-data.iloc[i-1]['recovered'])
data['daily_recovered']=daily_recovered
```

```python
fig7_1 = go.Figure()
fig7_1.add_trace(go.Scatter(x=data['date'], y=data['daily_cases'], mode='lines+markers', marker=dict(size=10), name='Daily Cases'))
fig7_1.add_trace(go.Scatter(x=data['date'], y=data['daily_recovered'], mode='lines+markers', marker=dict(size=10), name='Daily Recovered'))
```
